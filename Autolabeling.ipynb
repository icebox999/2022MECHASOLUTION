```python
from konlpy.tag import Mecab

mecab = Mecab()
```

```python
knu=pd.read_csv("/content/drive/MyDrive/data/감성사전.csv")
print(knu)
```

```python
pos = []
neg = []
neu = []
```

```python
for i in range(0, knu.shape[0]):
  row = knu.iloc[i]
  if row['polarity'] == 1:
    pos.append(row['word_root'])
  elif row['polarity'] == -1:
    neg.append(row['word_root'])
  else:
    neu.append(row['word_root'])

print(pos)
print(neg)
print(neu)
```

```python
class Tokenizer:
  def __init__(self, tagger):
    self.tagger = tagger
  def pos_tag(self, sentence):
    pos = self.tagger.pos(sentence)
    pos = sorted([f'{word}/{tag}' for word, tag in pos])
    return pos

mecab = Tokenizer(Mecab())
```

```python
print(df['댓글'][0])
```

```python
df['태그'] = 0
for i in range(len(df)):
  df['태그'][i] = mecab.pos_tag(df['댓글'][i])
print(df)
```

```python
def get_score_morphs(df, tag_colname, pos, neg):
  df["score_morph"] = 0
  df["label_morph"] = 0
  df["score"] = ""
  global posn
  global negn
  global neun
  posn = 0
  negn = 0
  neun = 0

  for i in range(len(df)):
    sentence = df[tag_colname][i]
    val = 0
    for morph in sentence:
      #print(morph)
      if morph in pos:
        val += 1
      elif morph in neg:
        val -= 1 
      print(val)

    df["score"][i] = val

    if val > 0:
      df["label_morph"][i] = "긍정"
      posn += 1
    elif val < 0:
      df["label_morph"][i] = "부정"
      negn+=1
    else:
      df["label_morph"][i] = "중립"
      neun+=1
    
  return df
```

```python
get_score_morphs(df, "태그", pos, neg)
```
